```python
transactions=pd.read_csv('../data/transactions/transactions.csv')
transactions.head(10)
```

```python
GBP=len(transactions[transactions['CURRENCY']=='GBP'])
print(GBP)
```

```python
NUSD=len(transactions[transactions['CURRENCY']!='USD'])
print(NUSD)
```

```python
ave=np.mean(transactions[transactions['CURRENCY'] == 'USD'].AMOUNT)
me=np.median(transactions[transactions['CURRENCY'] == 'USD'].AMOUNT)
print(ave,'\n',me)
```

```python
currencys = transactions['CURRENCY'].unique()
numbers = [len(transactions[transactions['CURRENCY']==i]) for i in currencys]
data = {'currency':currencys,'numbers':numbers}
table = pd.DataFrame(data)
print(table)
```
